```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
res = pd.read_csv('result_dengan_label.csv')
print(res.dtypes)
sol = {}
for i in range(1,5):
    print(i)
    sol.update({i:pd.read_csv('sol{}_dengan_label.csv'.format(i))})
    print(sol[i].dtypes)
```

```python
dict_rata2 = {'nomor fungsi fx':[1,2,3,4], 'rata-rata jumlah iterasi':[]}
for i in range(1,5):
    dict_rata2['rata-rata jumlah iterasi'].append(res[res['nomor fungsi fx'] == i].mean()['jumlah iterasi'])
pd.DataFrame(dict_rata2).plot(x='nomor fungsi fx')
plt.show()
```
